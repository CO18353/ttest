##  One-sample T-test

- The test will tell us whether means of the sample and the population are different.
- df= n-1


### Problem Statement
A producer claims that his bulbs have a mean life of 22 months. We randomly select 12 samples and check their life.

x = [21.5, 24.5, 18.5, 17.2, 14.5, 23.2, 22.1, 20.5, 19.4, 18.1, 24.1, 18.5]

Can you regard the producer's claim at 1% level of confidence?

```python
from scipy.stats
import numpy as np
x =  [21.5, 24.5, 18.5, 17.2, 14.5, 23.2, 22.1, 20.5, 19.4, 18.1, 24.1, 18.5]
x = sorted(x)
mean=round(np.mean(x),2)
sd=round(np.std(x),2)
print("Mean: ",mean," Std Dev: ",sd)
plt.plot(x, norm.pdf(x, mean, sd))
plt.show()
```

```python
from scipy.stats import ttest_1samp, t, norm
import matplotlib.pyplot as plt
import numpy as np

df = 11        # Degrees of freedom
alpha = 0.01    # Significence level
ntails = 2      # Number of tails 

## Calculate critical t-score
tcrit = abs(t.ppf(alpha/ntails, df))
tvalue, pvalue = ttest_1samp(x, popmean=22)
print("T value: ", tvalue)
print("Critical value: ",tcrit)
plt.figure()
xs = np.linspace(-5,5,1000)
plt.plot(xs, t.pdf(xs,df), 'k', label="T-Distribution PDF for df=11")
plt.plot(tvalue,t.pdf(tvalue,df),"bo", label="t value")
plt.vlines([-tcrit, tcrit], 0.0, t.pdf(tcrit,df), colors='r', label="t critical")
plt.legend()
```
